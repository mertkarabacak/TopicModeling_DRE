```python
import pandas as pd
```

```python
#Read data.

data = pd.read_csv("/content/drive/MyDrive/Topic Modeling-DRE/scopus.csv", index_col = None)
data.head()
```

```python
#Remove unwanted columns.

columns = ['Title', 'Year', 'Source title', 'Cited by', 'Abstract']

data = data[columns]

data.shape
```

```python
data = data[data['Abstract'] != '[No abstract available]']

data.shape
```

```python
#Create a 'Title + Abstract' column.

data['Title + Abstract'] = 'Title: ' + data['Title'] + '\n\n' + 'Abstract: ' + data['Abstract']
```

```python
#Bin the "Cited by" column into four quartiles and turn it into categorical data.

data['Citation Quartiles'] = pd.qcut(data['Cited by'], q=4, labels=['Q4', 'Q3', 'Q2', 'Q1'])
```

```python
#Recategorize journals.

top_10_sources = data['Source title'].value_counts().nlargest(10).index
data['Journal'] = data['Source title'].apply(lambda x: x if x in top_10_sources else 'Other')
```

```python
#Save data.

data.to_csv('/content/drive/MyDrive/Topic Modeling-DRE/data.csv')
```
